```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Assume results is your Results object

# Load the original image
image = results[0].orig_img

# Iterate over detected boxes
for box in results[0].boxes:
    # Box format: (x1, y1, x2, y2, confidence, class_id)
    cls_id = results[0].boxes.cls.item()
    conf = results[0].boxes.conf.item()

    x1, y1, x2, y2 = box.xyxy[0]
    class_name = results[0].names[int(cls_id)]
    label = f'{class_name} {conf:.2f}'

    # Draw bounding box
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
    # Put label
    cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
print(results[0].names[cls_id])
# Convert image to RGB (from BGR) and display using matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
import os
import cv2
directory_path = "/kaggle/working/datasets/affectnet-yolo-format/YOLO_format/test/images/"

all_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
def preprocess_image(image_path, image_size):
    image = Image.open(image_path)
    
    image = image.resize(image_size)
    image = np.array(image)
    image = image.astype(np.float32) / 255.0  
    image = np.transpose(image, (2, 0, 1)) 
    return image

image_size = (40, 40)  
batch_size = 16  

new_im = Image.new('RGB', (640, 640))
image_paths = all_files[:batch_size]
imaws = [preprocess_image(image_path, image_size) for image_path in image_paths]
input_data = np.stack([preprocess_image(image_path, image_size) for image_path in image_paths])
# x_offset = 0
print(len(imaws))
matrix = imaws[0].reshape(4, 4, 640, 640, 3)
horizontal_concatenated = np.concatenate(matrix, axis=1)

# Concatenate vertically (along rows)
final_image = np.concatenate(horizontal_concatenated, axis=0)
# Display the concatenated image
plt.imshow(final_image)
plt.axis('off')  # Hide axes
plt.show()

# plt.imshow(new_im)
```
